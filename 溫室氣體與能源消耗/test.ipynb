```python
import matplotlib.pyplot as plt

# 假設這裡是你的數據，能源消耗量和溫室氣體排放量的列表
energy_consumption = [100, 120, 80, 150, 90]  # 能源消耗量，例如每年的能源消耗量（單位可以是任意的，例如TBtu）
ghg_emissions = [50, 60, 40, 70, 45]  # 溫室氣體排放量，例如每年的二氧化碳排放量（單位可以是任意的，例如Mt CO2）

# 假設這裡是你的年份或地區的標籤
years = [2010, 2011, 2012, 2013, 2014]  # 年份，例如每年的數據對應的年份

# 繪製散點圖
plt.figure(figsize=(8, 6))  # 設置圖的大小（寬度為8英寸，高度為6英寸）
plt.scatter(energy_consumption, ghg_emissions, marker='o', c='blue', alpha=0.75)  # 繪製散點圖，指定顏色和透明度
plt.title('Energy Consumption vs. GHG Emissions')  # 設置圖的標題
plt.xlabel('Energy Consumption (TBtu)')  # 設置橫軸標籤
plt.ylabel('GHG Emissions (Mt CO2)')  # 設置縱軸標籤

# 添加每個點的標籤，這裡以年份作為標籤
for i, year in enumerate(years):
    plt.annotate(year, (energy_consumption[i], ghg_emissions[i]), textcoords="offset points", xytext=(5, 5), ha='center')

plt.grid(True)  # 添加網格線
plt.tight_layout()  # 自動調整布局，避免部分標籤被裁切
plt.show()  # 顯示圖形
```
